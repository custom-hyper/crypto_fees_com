# API Extraction

```python
#@title JSON Fees Extraction { display-mode: "code" }
import requests
import datetime
import time
import json
from google.colab import data_table

import plotly.graph_objects as go
import pandas as pd
import matplotlib.pyplot as plt
from pandas.io.json import json_normalize

#General Information table
import ast
#General Information table
response = requests.get("https://cryptofees.info/api/v1/fees")
output = response.text
json_parsed = json.loads(output)

try:
  df = pd.json_normalize(json_parsed, record_path =['protocols'])

except Exception:
    pass



df1 = pd.concat([pd.json_normalize(x) for x in df['fees'].values.tolist()], keys= df['id'])


df1.loc[:, ['date', 'fee']]

df1.loc[:'eth',].iloc[1:7]

df1['sum_fees']  =  df1.loc[:'eth','fee'].sum()
df1['sum_fees']  = df1['fee'].sum()

df1
```

```python
#@title JSON Extraction { display-mode: "code" }
import requests
import datetime
import time
import json
from google.colab import data_table

import plotly.graph_objects as go
import pandas as pd
import matplotlib.pyplot as plt



#General Information table
response = requests.get("https://cryptofees.info/api/v1/protocols")
output = response.text
json_parsed = json.loads(output)

full_talbe = pd.json_normalize(json_parsed, record_path =['protocols'])
full_talbe['name_merge'] = full_talbe['name'].str.replace(' ','')
full_talbe['id'] = full_talbe['tokenTicker']  + '.' + full_talbe['name_merge']

full_talbe.info()
full_talbe
```
